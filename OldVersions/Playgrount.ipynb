```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def show_image_differences(original_path, corrupt_path):
    try:
        # Load images
        original = cv2.imread(original_path)
        corrupt = cv2.imread(corrupt_path)

        if original is None or corrupt is None:
            print("Error: Could not load one or both images.")
            return

        if original.shape != corrupt.shape:
            print("Error: Image dimensions do not match.")
            return

        # Convert to grayscale
        gray_original = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
        gray_corrupt = cv2.cvtColor(corrupt, cv2.COLOR_BGR2GRAY)

        # Compute SSIM (Structural Similarity Index)
        score, diff = ssim(gray_original, gray_corrupt, full=True)
        diff = (diff * 255).astype(np.uint8)

        # Threshold the difference to highlight changes
        _, threshold_diff = cv2.threshold(diff, 30, 255, cv2.THRESH_BINARY)
        highlight = original.copy()
        highlight[threshold_diff > 0] = [0, 0, 255]  # Mark differences in red

        print(f"Similarity Score: {score:.2f}")

        # Show the images
        cv2.imshow("Approved Image", original)
        cv2.imshow("Corrupt Image", corrupt)
        cv2.imshow("Differences Highlighted", highlight)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

    except Exception as e:
        print(f"Error: {e}")

# ✅ Replace these paths with your image file paths
approved_image_path = "approved.jpg"
corrupt_image_path = "corrupt.jpg"

show_image_differences(approved_image_path, corrupt_image_path)
```

```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim
from PIL import Image

def load_tif_image(image_path):
    """Loads a .tif image and converts it to OpenCV format (BGR)."""
    try:
        image = Image.open(image_path)
        image = image.convert("RGB")  # Ensure it's in RGB mode
        return cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
    except Exception as e:
        print(f"Error loading {image_path}: {e}")
        return None

def show_image_differences(original_path, corrupt_path):
    try:
        # Load .tif images properly
        original = load_tif_image(original_path)
        corrupt = load_tif_image(corrupt_path)

        if original is None or corrupt is None:
            print("Error: Could not load one or both images.")
            return

        if original.shape != corrupt.shape:
            print("Error: Image dimensions do not match.")
            return

        # Convert to grayscale
        gray_original = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
        gray_corrupt = cv2.cvtColor(corrupt, cv2.COLOR_BGR2GRAY)

        # Compute SSIM (Structural Similarity Index)
        score, diff = ssim(gray_original, gray_corrupt, full=True)
        diff = (diff * 255).astype(np.uint8)

        # Threshold the difference to highlight changes
        _, threshold_diff = cv2.threshold(diff, 30, 255, cv2.THRESH_BINARY)
        highlight = original.copy()
        highlight[threshold_diff > 0] = [0, 0, 255]  # Mark differences in red

        print(f"Similarity Score: {score:.2f}")

        # Show the images
        cv2.imshow("Approved Image", original)
        cv2.imshow("Corrupt Image", corrupt)
        cv2.imshow("Differences Highlighted", highlight)

        # Wait for a key press to close the images
        cv2.waitKey(0)  # This will wait for any key press
        cv2.destroyAllWindows()  # Close all OpenCV windows after key press

    except Exception as e:
        print(f"Error: {e}")

# ✅ Replace these paths with your .tif image file paths
approved_image_path = "aa.tif"
corrupt_image_path = "cc.tif"

show_image_differences(approved_image_path, corrupt_image_path)
```
